```python

        

@lview.parallel(block=True)
def one_compute_mechanism(args):
    import sys
    sys.path.append('..')
    import studies.efficiency as e; reload(e)
    import strategic_trust_models as stm; reload(stm)
    
    def avg_rank_change(scores, agents, is_global):
        if not agents:
            return 0
        
        from scipy import stats
        import numpy as np
        def rank_change(score_vector, agent):
            rank = stats.rankdata(score_vector)
            return rank[agent] - agent - 1

        if is_global:
            return np.mean([rank_change(scores, agent) for agent in agents])
        else:
            return np.mean([
                np.mean([rank_change(score, agent) for agent in agents])
                for score in scores])
    
    MECHANISMS = [
    (True, 'global_pagerank', stm.global_pagerank),
    (False, 'person_pagerank', stm.person_pagerank),
    (True, 'global_hitting_time', stm.global_hitting_time),
    (False, 'person_hitting_time', stm.person_hitting_time),
    (False, 'person_max_flow', stm.person_max_flow),
    (False, 'person_shortest_path', stm.person_shortest_path),
]

    graph, mech_index, num_strategic, sybil_pct, cutlinks, gensybils = args
    
    is_global, name, func = MECHANISMS[mech_index]
    scores, strategic_agents = func(graph, num_strategic, sybil_pct, cutlinks, gensybils, return_strategic_agents=True)
    nonstrategic_agents = list(set(range(len(graph))) - set(strategic_agents))
    
    return {
        'info': e.compute_informativeness(graph.agent_types, scores, is_global),
        'eff': e.compute_efficiency(graph.agent_types, scores, is_global),
        'strat_rank_change': avg_rank_change(scores, strategic_agents, is_global),
        'nonstrat_rank_change': avg_rank_change(scores, nonstrategic_agents, is_global),
    }

@lview.parallel(block=True)
def one_gen_graph(args):
    import sys
    sys.path.append('..')
    from trust_graph import TrustGraph
    return TrustGraph(*args)
```

```python
MECHANISMS = [
    (True, 'global_pagerank', stm.global_pagerank),
    (False, 'person_pagerank', stm.person_pagerank),
    (True, 'global_hitting_time', stm.global_hitting_time),
    (False, 'person_hitting_time', stm.person_hitting_time),
    (False, 'person_max_flow', stm.person_max_flow),
    (False, 'person_shortest_path', stm.person_shortest_path),
]
```

```python
@lview.parallel(block=True)
def one_compute_mechanism(args):
    import sys
    sys.path.append('..')
    import studies.efficiency as e; reload(e)
    import strategic_trust_models as stm; reload(stm)
    
    MECHANISMS = [
    (True, 'global_pagerank', stm.global_pagerank),
    (False, 'person_pagerank', stm.person_pagerank),
    (True, 'global_hitting_time', stm.global_hitting_time),
    (False, 'person_hitting_time', stm.person_hitting_time),
    (False, 'person_max_flow', stm.person_max_flow),
    (False, 'person_shortest_path', stm.person_shortest_path),
]

    graph, mech_index, num_strategic, sybil_pct, cutlinks, gensybils = args
    
    is_global, name, func = MECHANISMS[mech_index]
    scores = func(graph, num_strategic, sybil_pct, cutlinks, gensybils)
    
    return {
        'info': e.compute_informativeness(graph.agent_types, scores, is_global),
        'eff': e.compute_efficiency(graph.agent_types, scores, is_global)
    }

@lview.parallel(block=True)
def one_gen_graph(args):
    import sys
    sys.path.append('..')
    from trust_graph import TrustGraph
    return TrustGraph(*args)
```

```python
# dim0: mechanisms
# dim1: strategic_counts
# dim2: num_iters
NUM_NODES = 50
NUM_EDGES = 30
NUM_SAMPLES = 8
NUM_STRATEGIC = 15

def info_by_strategic_counts(strat_pcts, cutlinks, gensybils, num_iters,
                             sybil_pct=0.10, agent_type_prior='uniform', edge_strategy='uniform', edge_weight_strategy='prior'):
    g_args = [(NUM_NODES, agent_type_prior, edge_strategy, NUM_EDGES, edge_weight_strategy, NUM_SAMPLES) for _ in xrange(num_iters)]
    graphs = one_gen_graph.map(g_args)
    num_strategic = [int(sp * NUM_NODES / 100.0) for sp in strat_pcts]
    args = [[[[g, i, sc, sybil_pct, cutlinks, gensybils]
              for g in graphs] for sc in num_strategic] for i in xrange(len(MECHANISMS))]
    flattened_args = [val for arr2 in args for arr in arr2 for val in arr]
    
#     g_args = [[(NUM_NODES, agent_type_prior, edge_strategy, NUM_EDGES, edge_weight_strategy, NUM_SAMPLES)
#                for _ in sybil_pcts] for _ in xrange(num_iters)]
#     ft_g_args = [val for arr in g_args for val in arr]
#     graphs = one_gen_graph.map(ft_g_args)
#     args = [[[g[len(strategic_counts)
#     args = [[[[g, i, sc, sybil_pct, cutlinks, gensybils]
#               for g in graphs] for sc in strategic_counts] for i in xrange(len(e.MECHANISMS))]
#     flattened_args = [val for arr2 in args for arr in arr2 for val in arr]
    
    results = one_compute_mechanism.map(flattened_args)
#     results = map(one_compute_mechanism.func, flattened_args)
    
    infos = np.array(map(op.itemgetter('info'), results)).reshape((len(MECHANISMS), len(strat_pcts), num_iters))
    effs = np.array(map(op.itemgetter('eff'), results)).reshape((len(MECHANISMS), len(strat_pcts), num_iters))
    
    return {
        'strat_pcts': strat_pcts,
        'num_strategic': num_strategic,
        'cutlinks': cutlinks,
        'gensybils': gensybils,
        'num_iters': num_iters,
        'sybil_pct': sybil_pct,
        'agent_type_prior': agent_type_prior,
        'edge_strategy': edge_strategy,
        'edge_weight_strategy': edge_weight_strategy,
        'infos': infos,
        'effs': effs,
        'num_nodes': NUM_NODES,
        'num_edges': NUM_EDGES,
        'num_samples': NUM_SAMPLES
    }

MARKERS = ['o', 's', '^', 'v', '*', 'p']
LABELS = {
    'global_pagerank': 'Global PageRank',
    'person_pagerank': 'Person. PageRank',
    'global_hitting_time': 'Global Hitting Time',
    'person_hitting_time': 'Person. Hitting Time',
    'person_max_flow': 'Person. Max Flow',
    'person_shortest_path': 'Person. Shortest Path'
}

def plot_by_strategic_counts(strat_pcts, num_strategic, cutlinks, gensybils, num_iters, sybil_pct, agent_type_prior, edge_strategy, edge_weight_strategy, infos, effs, **data):
    
    max_val = 10
    max_index = np.where(np.array(strat_pcts) > max_val)[0][0]
    xticks = [0, 20, 40, 60, 80]
    zoom_xticks = [0, 2, 4, 6, 8, 10]
    xt_lambda = lambda x: "%d%% (%d)" % (x, int(x * data['num_nodes'] / 100.0))
    
    for i, (_, name, _) in enumerate(MECHANISMS):
        info = infos[i, :, :].mean(axis=1)
        plt.plot(strat_pcts, info, MARKERS[i] + '--', label=LABELS[name])
    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
#     plt.xticks(strategic_counts, strategic_counts)
    plt.xticks(xticks, map(xt_lambda, xticks))
    plt.xlabel('% of agents that are strategic (# of strategic agents)')
    plt.ylabel('Informativeness')
    plt.title('Informativeness of Reputation Mechanisms\n'
                 "G(%d, %d) '%s' prior, '%s' edges, '%s' weights, %d samples\n"
                 '%s cutting outlinks, %s sybils, %0.2f sybils (%d iters)' % (
                        data['num_nodes'], data['num_edges'],
                        agent_type_prior, edge_strategy, edge_weight_strategy, data['num_samples'],
                        "with" if cutlinks else "without", "with" if gensybils else "without", sybil_pct, num_iters),
                 fontsize=18)
    plt.show()
    
    # Zoom in
    for i, (_, name, _) in enumerate(MECHANISMS):
        info = infos[i, :, :].mean(axis=1)
        plt.plot(strat_pcts[:max_index], info[:max_index], MARKERS[i] + '--', label=LABELS[name])
    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.xticks(zoom_xticks, map(xt_lambda, zoom_xticks))
    plt.xlabel('% of agents that are strategic (# of strategic agents)')
    plt.ylabel('Informativeness')
    plt.title('Informativeness of Reputation Mechanisms\n'
                 "G(%d, %d) '%s' prior, '%s' edges, '%s' weights, %d samples\n"
                 '%s cutting outlinks, %s sybils, %0.2f sybils (%d iters)' % (
                        data['num_nodes'], data['num_edges'],
                        agent_type_prior, edge_strategy, edge_weight_strategy, data['num_samples'],
                        "with" if cutlinks else "without", "with" if gensybils else "without", sybil_pct, num_iters),
                 fontsize=18)
    plt.show()
    
    
#     Efficiency
    for i, (_, name, _) in enumerate(MECHANISMS):
        eff = effs[i, :, :].mean(axis=1)
        plt.plot(strat_pcts, eff, MARKERS[i] + '--', label=LABELS[name])
    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.xticks(xticks, map(xt_lambda, xticks))
    plt.xlabel('% of agents that are strategic (# of strategic agents)')
    plt.ylabel('Efficiency')
    plt.title('Efficiency of Reputation Mechanisms\n'
                 "G(%d, %d) '%s' prior, '%s' edges, '%s' weights, %d samples\n"
                 '%s cutting outlinks, %s sybils, %0.2f sybils (%d iters)' % (
                        data['num_nodes'], data['num_edges'],
                        agent_type_prior, edge_strategy, edge_weight_strategy, data['num_samples'],
                        "with" if cutlinks else "without", "with" if gensybils else "without", sybil_pct, num_iters),
                 fontsize=18)
    plt.show()
    
    
    for i, (_, name, _) in enumerate(MECHANISMS):
        eff = effs[i, :, :].mean(axis=1)
        plt.plot(strat_pcts[:max_index], eff[:max_index], MARKERS[i] + '--', label=LABELS[name])
    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.xticks(zoom_xticks, map(xt_lambda, zoom_xticks))
    plt.xlabel('% of agents that are strategic (# of strategic agents)')
    plt.ylabel('Efficiency')
    plt.title('Efficiency of Reputation Mechanisms\n'
                 "G(%d, %d) '%s' prior, '%s' edges, '%s' weights, %d samples\n"
                 '%s cutting outlinks, %s sybils, %0.2f sybils (%d iters)' % (
                        data['num_nodes'], data['num_edges'],
                        agent_type_prior, edge_strategy, edge_weight_strategy, data['num_samples'],
                        "with" if cutlinks else "without", "with" if gensybils else "without", sybil_pct, num_iters),
                 fontsize=18)
    plt.show()
```

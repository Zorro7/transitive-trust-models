```python
from trust_graph import TrustGraph
from scipy import stats
import networkx as nx
import trust_models as tm
```

```python
gdense = TrustGraph(50, 'uniform', 'uniform', 29, 'noisy', 32)
```

```python
mfs = tm.max_flow(gdense)
mfs_jitter = mfs + np.random.normal(scale=0.0001, size=(50 * 50)).reshape(50, 50)
```

```python
plot([stats.spearmanr(gdense.agent_types, mfs_jitter[i, :])[0] for i in xrange(50)])
xlabel('Agent Ranking (higher ranking is higher type)')
ylabel('Spearman correlation against true types')
suptitle('Max flow accuracy by agent type in a dense graph (fully connected w/ 50 nodes)')
```

```python
plot(range(3, 49), [stats.spearmanr(gdense.agent_types[:i], mfs_jitter[i, :i])[0]
                    for i in xrange(3, 49)], label='Lower type agents')
plot(range(3, 47), [stats.spearmanr(gdense.agent_types[i+1:], mfs_jitter[i, i+1:])[0]
                    for i in xrange(3, 47)], label='Higher type agents')
legend(loc='best')
xlabel('Agent Ranking (higher ranking is higher type)')
ylabel('Spearman correlation')
suptitle('Max Flow accuracy, partitioned by high/low types, in connected graph of 50 nodes')
margins(0.05)
```

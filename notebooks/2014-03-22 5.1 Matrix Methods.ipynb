```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import hitting_time.mat_hitting_time as m
import operator as op
import time
```

```python
def one_LS_SL_HT(graph):
    start_time = time.clock()
    ht = m.personalized_LS_step_length_ht(graph)
    elapsed = time.clock() - start_time
    
    return {
        'hitting_times': ht,
        'runtime': elapsed
        }

def one_LS_prob_HT(graph):
    start_time = time.clock()
    ht = m.personalized_LS_prob_ht(graph)
    elapsed = time.clock() - start_time
    
    return {
        'hitting_times': ht,
        'runtime': elapsed
        }

def one_eigen_prob_HT(graph):
    start_time = time.clock()
    ht = m.personalized_eigen_prob_ht(graph)
    elapsed = time.clock() - start_time
    
    return {
        'hitting_times': ht,
        'runtime': elapsed
        }
```

```python
# 1. Look at running time by graph size
# We assume K = N/5
def runtime_by_size(sizes, num_iters):
#     graphs = [[nx.barabasi_albert_graph(size, size / 10) for _ in xrange(num_iters)] for size in sizes]
    graphs = [[nx.complete_graph(size) for _ in xrange(num_iters)] for size in sizes]
    flattened_graphs = [val for arr in graphs for val in arr]
    
    ls_sl_ht = map(one_LS_SL_HT, flattened_graphs); print 'Completed LS_step_length';
    ls_sl_runtimes = np.array(map(op.itemgetter('runtime'), ls_sl_ht)).reshape((len(sizes), num_iters))
    ls_prob_ht = map(one_LS_prob_HT, flattened_graphs); print 'Completed LS_prob';
    ls_prob_runtimes = np.array(map(op.itemgetter('runtime'), ls_prob_ht)).reshape((len(sizes), num_iters))
    eigen_prob_ht = map(one_eigen_prob_HT, flattened_graphs); print 'Completed eigen_prob';
    eigen_prob_runtimes = np.array(map(op.itemgetter('runtime'), eigen_prob_ht)).reshape((len(sizes), num_iters))
    
    return {
        'num_iters': num_iters,
        'sizes': sizes,
        'ls_sl': ls_sl_runtimes,
        'ls_prob': ls_prob_runtimes,
        'eigen_prob': eigen_prob_runtimes,
    }

LABELS = [
    ('ls_sl', 'LS Step Length', 'b', '*'),
    ('ls_prob', 'LS Probability', 'g', 'o'),
#     ('eigen_prob', 'Eigen Probability', 'r', 'v'),
]

def plot_runtimes_by_size(sizes, num_iters, **data):
    for key, label, color, marker in LABELS:
        rt = data[key].mean(axis=1)
        plt.plot(sizes, rt, marker + '--', label=label, c=color)
        
    plt.margins(0.07)
    plt.legend(loc='best')
    plt.xticks(sizes, sizes)
    plt.xlabel('Number of nodes')
    plt.ylabel('Running time (secs)')
    plt.suptitle('Running time of exact algorithms by graph size\n'
                 'Complete Graphs (%d iters)' % num_iters)
#                     'Barabasi-Albert graphs, K = N/10 (%d iters)' % num_iters)
    plt.show()
    
```

```python
d = runtime_by_size([10, 20, 30, 40, 50, 60, 70, 80], 10)
```

```python
plot_runtimes_by_size(**d)
```

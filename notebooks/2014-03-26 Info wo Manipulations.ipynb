```python
import numpy as np
import matplotlib.pyplot as plt
import studies.efficiency as e; reload(e);
import strategic_trust_models as stm; reload(stm);
from trust_graph import TrustGraph
import operator as op
import time
import math
```

```python
from IPython.parallel import Client
cli = Client()
lview = cli.load_balanced_view()
```

```python
@lview.parallel(block=True)
def one_compute_mechanism(args):
    import sys
    sys.path.append('..')
    import studies.efficiency as e
    from trust_graph import TrustGraph
    
#     graph = TrustGraph(*args[:6])
#     mech_index = args[-1]

    graph, mech_index = args
    
    is_global, name, func = e.MECHANISMS[mech_index]
    scores = func(graph, 0, 0, False, False) # No strategic, no sybils, no cutlinks, no sybils
    
    return {
        'info': e.compute_informativeness(graph.agent_types, scores, is_global),
        'eff': e.compute_efficiency(graph.agent_types, scores, is_global)
    }

@lview.parallel(block=True)
def one_gen_graph(args):
    import sys
    sys.path.append('..')
    from trust_graph import TrustGraph
    return TrustGraph(*args)
```

```python
# dim0: mechanisms
# dim1: edge_counts
# dim2: num_iters
def info_by_edge_count(edge_counts, num_iters, agent_type_prior='uniform', edge_strategy='uniform', edge_weight_strategy='noisy'):
    g_args = [[(50, agent_type_prior, edge_strategy, ec, edge_weight_strategy, 32) for _ in xrange(num_iters)] for ec in edge_counts]
    ft_graph_args = [val for arr in g_args for val in arr]
    graphs = one_gen_graph.map(ft_graph_args)
    print 'Finished generating graphs'
#     flattened_graphs = [val for arr in graphs for val in arr]
    args = [[[g, i] for g in graphs] for i in xrange(len(e.MECHANISMS))]
    flattened_args = [val for arr in args for val in arr]
#     args = [[[[50, 'uniform', 'uniform', ec, 'noisy', 32, i]
#                for _ in xrange(num_iters)]
#               for ec in edge_counts]
#              for i in xrange(len(e.MECHANISMS))]
#     flattened_args = [val for arr2 in args for arr in arr2 for val in arr]
    
    results = one_compute_mechanism.map(flattened_args)
    
    infos = np.array(map(op.itemgetter('info'), results)).reshape((len(e.MECHANISMS), len(edge_counts), num_iters))
    effs = np.array(map(op.itemgetter('eff'), results)).reshape((len(e.MECHANISMS), len(edge_counts), num_iters))
    
    return {
        'edge_counts': edge_counts,
        'num_iters': num_iters,
        'agent_type_prior': agent_type_prior,
        'edge_strategy': edge_strategy,
        'edge_weight_strategy': edge_weight_strategy,
        'infos': infos,
        'effs': effs
    }

def info_by_sample_count(sample_counts, num_iters, agent_type_prior='uniform', edge_strategy='uniform', edge_weight_strategy='noisy'):
    sample_counts.append(float('inf'))
    graphs = [[TrustGraph(50, agent_type_prior, edge_strategy, 15, edge_weight_strategy, sc) for _ in xrange(num_iters)] for sc in sample_counts]
    flattened_graphs = [val for arr in graphs for val in arr]
    args = [[[g, i] for g in flattened_graphs] for i in xrange(len(e.MECHANISMS))]
    flattened_args = [val for arr in args for val in arr]
#     args = [[[[50, 'uniform', 'uniform', ec, 'noisy', 32, i]
#                for _ in xrange(num_iters)]
#               for ec in edge_counts]
#              for i in xrange(len(e.MECHANISMS))]
#     flattened_args = [val for arr2 in args for arr in arr2 for val in arr]
    
    results = one_compute_mechanism.map(flattened_args)
    
    infos = np.array(map(op.itemgetter('info'), results)).reshape((len(e.MECHANISMS), len(sample_counts), num_iters))
    effs = np.array(map(op.itemgetter('eff'), results)).reshape((len(e.MECHANISMS), len(sample_counts), num_iters))
    
    return {
        'sample_counts': sample_counts,
        'num_iters': num_iters,
        'agent_type_prior': agent_type_prior,
        'edge_strategy': edge_strategy,
        'edge_weight_strategy': edge_weight_strategy,
        'infos': infos,
        'effs': effs
    }

MARKERS = ['o', 's', '^', 'v', '*', 'p']
    
def plot_by_edge_count(edge_counts, num_iters, agent_type_prior, edge_strategy, edge_weight_strategy, infos, effs):
    for i, (_, name, _) in enumerate(e.MECHANISMS):
        info = infos[i, :, :].mean(axis=1)
        plt.plot(edge_counts, info, MARKERS[i] + '--', label=name)
    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.xticks(edge_counts, edge_counts)
    plt.xlabel('Number of edges')
    plt.ylabel('Informativeness')
    plt.suptitle('Informativeness of Reputation Mechanisms\n'
                 "G(50, K) '%s' prior, '%s' edges, '%s' weights (%d iters)" % (
                        agent_type_prior, edge_strategy, edge_weight_strategy, num_iters))
    plt.show()
    

def plot_by_sample_count(sample_counts, num_iters, agent_type_prior, edge_strategy, edge_weight_strategy, infos, effs):
    log_sc = [math.log(x, 2) for x in sample_counts if x != float('inf')]
    xtick_locs = log_sc + [max(log_sc) + 2]
    xtick_lbls = map(str, log_sc) + ['inf']
    
    for i, (_, name, _) in enumerate(e.MECHANISMS):
        info = infos[i, :, :].mean(axis=1)
        plt.plot(xtick_locs, info, MARKERS[i] + '--', label=name)

    
    plt.margins(0.07)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.xticks(xtick_locs, xtick_lbls)
    plt.xlabel('Number of samples per edge')
    plt.ylabel('Informativeness')
    plt.suptitle('Informativeness of Reputation Mechanisms\n'
                 "G(50, 15) '%s' prior, '%s' edges, '%s' weights (%d iters)" % (
                        agent_type_prior, edge_strategy, edge_weight_strategy, num_iters))
    plt.show()
```
